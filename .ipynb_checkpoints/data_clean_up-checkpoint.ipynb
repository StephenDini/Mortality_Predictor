#### Literacy data is not available for every year so the most recent available year, which is what the file has, will be used
#### Religion is not available for all years. However, religious affiliation breakdown should not change rapidly so 2010 will be used. Every year past 2015 is a projection
#### Alchohol years to be used will be 2000, 2005, 2010, 2015
#### Mortality data is available for each of the alchohol years so those will be selected
#### Population growth data is available for each of the alchohol years so those will be selected
#### GDP is available for each of the alchohol years so those will be selected
#### GDP percapita is available for each of the alchohol years so those will be selected

```python
# Add a mortality_state column that bins the mortailty from 2015 into ctaegories of high, medium, and low numericly 3,2,1
high = 20
med = 10
# Turn mortality_rate_2015 into a %
agglomerated_df['mortality_rate'] = agglomerated_df['mortality_rate_2015'].div(1000).mul(100)

# Round data to 2 decimal places
# Syntax found at https://www.geeksforgeeks.org/python-pandas-dataframe-round/
agglomerated_df = agglomerated_df.round({"mortality_rate":2})


# Syntax to add column based on another column range found at
#https://stackoverflow.com/questions/59642338/creating-new-column-based-on-condition-on-other-column-in-pandas-dataframe
agglomerated_df['mortality_state']=np.select([agglomerated_df['mortality_rate']<med,agglomerated_df['mortality_rate'].between(med,high)],[1,2],3)
agglomerated_df.head(10)
```

## Write Data required for mapping to GeoJSON format

```python
# Syntax to generate a GeoJSON file from a data frame found at
# https://geoffboeing.com/2015/10/exporting-python-data-geojson/


# Define function to generate GEOJSON format
def df_to_geojson(df, properties, lat='Latitude', lon='Longitude'):
    geojson = {'type':'FeatureCollection', 'features':[]}
    for _, row in df.iterrows():
        feature = {'type':'Feature',
                   'properties':{},
                   'geometry':{'type':'Point',
                               'coordinates':[]}}
        feature['geometry']['coordinates'] = [row[lon],row[lat]]
        for prop in properties:
            feature['properties'][prop] = row[prop]
        geojson['features'].append(feature)
    return geojson
```

```python
columns = ['Country','Longitude','Latitude','Literacy_Rate', 'mortality_rate','mortality_state','gdp_per_capita_2015','alcohol_rate_2015']
geojson = df_to_geojson(agglomerated_df, columns)
```

```python
output_filename = './cleaned_data/map.geojson'

# Dump the geojson format data to a file

with open(output_filename, 'w') as output_file:
    json.dump(geojson, output_file, indent=2) 
```
