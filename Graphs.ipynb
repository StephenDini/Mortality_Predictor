# Alcohol Rate (2015) vs Mortality

```python
# Add Matplotlib inline magic command
%matplotlib inline

# Import dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
# read cleaned data CSV file
file_path = Path('./cleaned_data/data_by_country.csv')
df = pd.read_csv(file_path)
#df = pd.read_csv(file_path, index_col=0)
df.head(10)   
```

```python
# Filter the DataFrame down only to those columns to chart
am = df[["alcohol_rate_2015","mortality_rate"]]


am.head()
```

```python
#scatter plot
x = am["mortality_rate"]
y = am["alcohol_rate_2015"]

plt.figure(figsize=(20, 12))
plt.scatter(x, y, s=75)

# Incorporate the other graph properties
plt.title("Alcohol vs Mortality Rate (2015)", fontsize=20)
plt.ylabel("Alcohol Rate (%)", fontsize=14)
plt.xlabel("Mortality Rate(%)", fontsize=14)

# Add correlation line
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x, p(x), color='r')

# Save the figure.
plt.savefig("./docs/pictures/Alcohol_vs_mortality.png")
```

# GDP (2015) vs Mortality

```python
# Add Matplotlib inline magic command
%matplotlib inline

# Import dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
# read cleaned data CSV file
file_path = Path('./cleaned_data/data_by_country.csv')
df = pd.read_csv(file_path)
#df = pd.read_csv(file_path, index_col=0)
df.head(10)   
```

```python
# Filter the DataFrame down only to those columns to chart
gdp_m = df[["gpd_2015","mortality_rate"]]


gdp_m.head()
```

```python
#scatter plot
x = gdp_m["mortality_rate"]
y = gdp_m["gpd_2015"]

plt.figure(figsize=(20, 12))
plt.scatter(x, y, s=75)

# Incorporate the other graph properties
plt.title("GDP vs Mortality Rate (2015)", fontsize=24)
plt.ylabel("GDP ($)", fontsize=18)
plt.xlabel("Mortality Rate(%)", fontsize=18)


# https://queirozf.com/entries/matplotlib-examples-number-formatting-for-axes-labels
# Format the labels
current_values = plt.gca().get_yticks()
# using format string '{:.0f}' here but you can choose others
plt.gca().set_yticklabels(['${:,.0f}'.format(x) for x in current_values])


# Add correlation line
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x, p(x), color='r')

# Save the figure.
plt.savefig("./docs/pictures/GDP_vs_mortality.png")
```
